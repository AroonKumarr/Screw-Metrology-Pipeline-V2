# 🔩 Screw Metrology Pipeline — End-to-End Notebook

> **Measurement Target:** Screws (M3/M4/M5 hex-head machine screws)  
> **Goal:** Automatically measure the **real-world width (diameter) and length** of a screw in **millimetres** from a single photo, using instance segmentation and a reference ArUco marker.

---

## 📋 Pipeline Overview

| Step | Stage | Description |
|------|-------|-------------|
| **Step 1** | Foundational | Camera Calibration + Dataset Collection |
| **Step 2** | Intermediate | Mask R-CNN Training + Segmentation |
| **Step 3** | Advanced | Pixel-to-MM Measurement + Validation |

```
📷 Raw Image
     │
     ▼
🔧 Undistort (Camera Calibration)
     │
     ▼
🧠 Mask R-CNN (Instance Segmentation)
     │
     ▼
📏 ArUco Reference Detection (px/mm ratio)
     │
     ▼
📐 Metric Output: Width & Height in mm
```

# 🔵 STEP 1 — Foundational: Camera Calibration & Data Collection

> **Theory:** The smartphone camera lens introduces **radial distortion** — straight lines appear curved, and pixel-to-mm scale varies across the image. Calibration computes a **Camera Matrix K** and **distortion coefficients (k₁, k₂, p₁, p₂)** using a checkerboard of known geometry.
>
> The core equation solved is:  
> `s·[u, v, 1]ᵀ = K · [R | t] · [X, Y, Z, 1]ᵀ`

### Step 1 Expected Outcomes
- ✅ Calibrated camera with documented parameters  
- ✅ Clean labelled dataset ready for training  
- ✅ All images undistorted and verified

## 1.1 — Checkerboard Detection & Calibration

````python
from calibration.calibrate import find_checkerboard_corners, calibrate_camera, save_calibration, generate_calibration_report

# ─── Configuration ─────────────────────────────────────────────────────────
BOARD_SIZE  = (9, 6)   # inner corners (cols × rows)
SQUARE_SIZE = 25.0     # mm per square

# ─── List calibration images ───────────────────────────────────────────────
calib_images = sorted(CALIB_IMG_DIR.glob('*.jpg')) + sorted(CALIB_IMG_DIR.glob('*.JPG')) + sorted(CALIB_IMG_DIR.glob('*.jpeg'))
print(f'📷 Found {len(calib_images)} calibration images in {CALIB_IMG_DIR}')

# ─── Test corner detection on first image ──────────────────────────────────
if calib_images:
    success, corners, img = find_checkerboard_corners(str(calib_images[0]), BOARD_SIZE)
    print(f'\n🔍 Corner detection on "{calib_images[0].name}": {"✅ FOUND" if success else "❌ NOT FOUND"}')
    if success:
        print(f'   Corners detected: {corners.shape[0]} points')
        # Draw corners
        vis = cv2.drawChessboardCorners(img.copy(), BOARD_SIZE, corners, success)
        vis_rgb = cv2.cvtColor(vis, cv2.COLOR_BGR2RGB)
        small = cv2.resize(vis_rgb, (0, 0), fx=0.2, fy=0.2)
        plt.figure(figsize=(10, 7))
        plt.imshow(small)
        plt.title(f'Detected Checkerboard Corners — {calib_images[0].name}', fontweight='bold')
        plt.axis('off')
        plt.tight_layout()
        plt.savefig(OUTPUTS_DIR / 'step1_corner_detection.png', dpi=120)
        plt.show()
else:
    print('⚠️  No calibration images found. Place checkerboard JPGs in:', CALIB_IMG_DIR)
````

````python
# ─── Run full calibration ─────────────────────────────────────────────────
print('🔧 Running camera calibration...')
print(f'   Board size : {BOARD_SIZE[0]} × {BOARD_SIZE[1]} inner corners')
print(f'   Square size: {SQUARE_SIZE} mm')
print()

cam_matrix, dist_coeffs, rvecs, tvecs, rms_error, img_size = calibrate_camera(
    image_dir=str(CALIB_IMG_DIR),
    board_size=BOARD_SIZE,
    square_size=SQUARE_SIZE
)

if cam_matrix is not None:
    print(f'\n✅ Calibration complete!')
    print(f'   RMS Reprojection Error : {rms_error:.4f} px')
    print(f'   Image size             : {img_size[0]} × {img_size[1]}')
    print(f'   Focal length (fx, fy)  : ({cam_matrix[0,0]:.1f}, {cam_matrix[1,1]:.1f})')
    print(f'   Principal point (cx,cy): ({cam_matrix[0,2]:.1f}, {cam_matrix[1,2]:.1f})')
    print(f'   Distortion (k1,k2,p1,p2): ({dist_coeffs[0,0]:.4f}, {dist_coeffs[0,1]:.4f}, {dist_coeffs[0,2]:.4f}, {dist_coeffs[0,3]:.4f})')
else:
    print('⚠️  Calibration failed — loading pre-saved calibration instead')
    cam_matrix  = np.load(str(CALIB_OUTPUT_DIR / 'camera_matrix.npy'))
    dist_coeffs = np.load(str(CALIB_OUTPUT_DIR / 'dist_coeffs.npy'))
    print('   Loaded from:', CALIB_OUTPUT_DIR)
````

````python
# ─── Count images across splits ───────────────────────────────────────────
def count_images(directory):
    exts = ['*.jpg','*.JPG','*.jpeg','*.png','*.PNG']
    total = 0
    for ext in exts:
        total += len(list(Path(directory).glob(ext)))
    return total

train_count  = count_images(DATASET_DIR / 'train' / 'images')
val_count    = count_images(DATASET_DIR / 'val'   / 'images') if (DATASET_DIR/'val'/'images').exists() else 0
test_count   = count_images(DATASET_DIR / 'test'  / 'images') if (DATASET_DIR/'test'/'images').exists() else 0

# ─── Load annotation stats ────────────────────────────────────────────────
ann_stats = {}
if TRAIN_ANN.exists():
    with open(TRAIN_ANN) as f:
        train_coco = json.load(f)
    ann_stats['train'] = {'images': len(train_coco.get('images',[])), 'annotations': len(train_coco.get('annotations',[]))}

if VAL_ANN.exists():
    with open(VAL_ANN) as f:
        val_coco = json.load(f)
    ann_stats['val'] = {'images': len(val_coco.get('images',[])), 'annotations': len(val_coco.get('annotations',[]))}

# ─── Print summary table ──────────────────────────────────────────────────
print('=' * 50)
print('       📊 DATASET SUMMARY')
print('=' * 50)
print(f'  Measurement target : Screws (hex-head)')
print(f'  Label format       : COCO JSON (polygon masks)')
print(f'  Classes            : background (0), screw (1)')
print()
print(f'  {"Split":<10} {"Images":<10} {"Annotations"}')
print(f'  {"-"*35}')
for split, stats in ann_stats.items():
    print(f'  {split:<10} {stats["images"]:<10} {stats["annotations"]}')
print(f'  {"-"*35}')
total_imgs = sum(s['images'] for s in ann_stats.values())
total_anns = sum(s['annotations'] for s in ann_stats.values())
print(f'  {"TOTAL":<10} {total_imgs:<10} {total_anns}')
print('=' * 50)
````

````python
# ─── Visualise sample annotated images from dataset ───────────────────────
import random

def show_coco_sample(ann_file, img_root, n=4, title='Annotated Samples'):
    if not Path(ann_file).exists():
        print(f'⚠️  Annotation file not found: {ann_file}'); return
    with open(ann_file) as f:
        coco = json.load(f)
    
    # Build id→filename map
    id2file = {img['id']: img['file_name'] for img in coco['images']}
    id2anns = {}
    for ann in coco['annotations']:
        id2anns.setdefault(ann['image_id'], []).append(ann)
    
    sample_ids = random.sample(list(id2file.keys()), min(n, len(id2file)))
    fig, axes = plt.subplots(1, len(sample_ids), figsize=(5*len(sample_ids), 5))
    if len(sample_ids) == 1: axes = [axes]
    
    for ax, img_id in zip(axes, sample_ids):
        fname = id2file[img_id]
        img_path = Path(img_root) / fname
        if not img_path.exists():
            # Try without directory prefix
            img_path = Path(img_root) / Path(fname).name
        
        if img_path.exists():
            img = np.array(Image.open(img_path).convert('RGB'))
            ax.imshow(img)
            # Draw bounding boxes
            for ann in id2anns.get(img_id, []):
                bb = ann['bbox']  # [x, y, w, h]
                rect = patches.Rectangle((bb[0],bb[1]), bb[2], bb[3],
                                          linewidth=2, edgecolor='lime', facecolor='none')
                ax.add_patch(rect)
                ax.text(bb[0], bb[1]-5, 'screw', color='lime', fontsize=8, fontweight='bold',
                        bbox=dict(facecolor='black', alpha=0.5, pad=1))
            ax.set_title(Path(fname).name[:25], fontsize=8)
        else:
            ax.text(0.5, 0.5, f'Image not found:\n{Path(fname).name}', ha='center', va='center', transform=ax.transAxes)
        ax.axis('off')
    
    fig.suptitle(title, fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'step1_dataset_samples.png', dpi=120)
    plt.show()

show_coco_sample(TRAIN_ANN, DATASET_DIR/'train'/'images', n=4, title='Step 1 — Labelled Training Samples (COCO format)')
````

# 🟡 STEP 2 — Intermediate: Model Training & Segmentation

> **Theory:** Mask R-CNN is a two-stage detector:  
> 1. **ResNet-50 + FPN backbone** extracts multi-scale feature maps  
> 2. **Region Proposal Network (RPN)** proposes candidate bounding boxes  
> 3. **RoI Align** crops and aligns features per proposal  
> 4. **Box head** refines box + classifies as *screw* or *background*  
> 5. **Mask head** (fully convolutional) generates a **28×28 pixel mask** per detected instance

### Step 2 Expected Outcomes
- ✅ Trained model with documented performance metrics  
- ✅ Reproducible training pipeline (config + this notebook)  
- ✅ Working inference on new images from calibrated camera

## 2.3 — Training Loss Curves

````python
# ─── Load training log ────────────────────────────────────────────────────
log_path = Path(TRAIN_CONFIG['output_dir']) / 'training_log.json'

if log_path.exists():
    with open(log_path) as f:
        history = json.load(f)
    
    epochs      = list(range(1, len(history['train_loss']) + 1))
    train_loss  = history['train_loss']
    val_loss    = history['val_loss']
    lr_values   = history.get('lr', [])
    
    best_epoch  = val_loss.index(min(val_loss)) + 1
    best_loss   = min(val_loss)
    
    # ─── Plot ─────────────────────────────────────────────────────────────
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
    
    # Loss curves
    ax1.plot(epochs, train_loss, 'o-', color='#2196F3', label='Train Loss', linewidth=2, markersize=5)
    ax1.plot(epochs, val_loss,   's-', color='#FF5722', label='Val Loss',   linewidth=2, markersize=5)
    ax1.axvline(x=best_epoch, color='#4CAF50', linestyle='--', alpha=0.8, label=f'Best Val (Ep {best_epoch})')
    ax1.scatter([best_epoch], [best_loss], color='#4CAF50', s=120, zorder=5)
    ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss')
    ax1.set_title('📉 Training & Validation Loss Curves', fontweight='bold')
    ax1.legend(); ax1.grid(True, alpha=0.3)
    
    # LR schedule
    if lr_values:
        ax2.plot(epochs, lr_values, '^-', color='#9C27B0', linewidth=2, markersize=5)
        ax2.set_xlabel('Epoch'); ax2.set_ylabel('Learning Rate')
        ax2.set_title('📊 Cosine Annealing LR Schedule', fontweight='bold')
        ax2.grid(True, alpha=0.3)
        ax2.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    
    plt.tight_layout()
    plt.savefig(OUTPUTS_DIR / 'step2_loss_curves.png', dpi=150)
    plt.show()
    
    # ─── Print metrics table ──────────────────────────────────────────────
    print('=' * 55)
    print('       📊 TRAINING PERFORMANCE METRICS')
    print('=' * 55)
    print(f'  Total epochs trained : {len(epochs)}')
    print(f'  Initial train loss   : {train_loss[0]:.4f}')
    print(f'  Final train loss     : {train_loss[-1]:.4f}')
    print(f'  Best val loss        : {best_loss:.4f} (Epoch {best_epoch})')
    print(f'  Loss improvement     : {((train_loss[0]-train_loss[-1])/train_loss[0]*100):.1f}%')
    print(f'  Model size           : {MODEL_WEIGHTS.stat().st_size/1e6:.1f} MB' if MODEL_WEIGHTS.exists() else '')
    print('=' * 55)
else:
    print(f'⚠️  Training log not found at: {log_path}')
````

# 🟢 STEP 3 — Advanced: Pixel-to-MM Measurement

> **Theory:** After obtaining a pixel-perfect mask, we need a **scale reference**.  
> We place an **ArUco marker** (known physical size = 198 mm) in the same plane as the screw.
>
> The conversion is:
> ```
> pixels_per_mm = marker_side_pixels / marker_side_mm
> object_mm     = object_pixels / pixels_per_mm
> ```
>
> `cv2.minAreaRect` gives us the **rotation-invariant** bounding box of the screw contour,  
> which correctly extracts **diameter (width)** and **length (height)** regardless of orientation.

### Step 3 — Expected Outcomes
- ✅ Accurate pixel-to-mm measurement with documented error analysis  
- ✅ End-to-end pipeline from raw image to metric output  
- ✅ Comprehensive accuracy validation against physical measurements

## 3.1 — ArUco Marker Detection & px/mm Ratio

## 3.4 — Calibration Dependency Explanation

````python
# ─── Demonstrate WHY calibration is critical for metrology ────────────────
print('=' * 60)
print('  🔬 WHY CAMERA CALIBRATION IS MANDATORY FOR METROLOGY')
print('=' * 60)
print('''
  Without calibration, the px/mm ratio is DIFFERENT across
  the image due to radial distortion:

    • Centre of image : 1 px ≈ 0.20 mm  (relatively accurate)
    • Edge of image   : 1 px ≈ 0.17 mm  (stretched by lens)

  If the screw is near an edge and the ArUco is at the centre,
  the calculated size will be WRONG by up to 15–20%.

  After undistortion:
    • Centre of image : 1 px ≈ 0.20 mm  ✅ uniform
    • Edge of image   : 1 px ≈ 0.20 mm  ✅ uniform

  This is why we ALWAYS call cv2.undistort() before measuring.
''')

# Visualise distortion effect numerically
k1, k2 = dist_coeffs[0,0], dist_coeffs[0,1]
cx_val  = cam_matrix[0,2]
cy_val  = cam_matrix[1,2]
fx_val  = cam_matrix[0,0]

print(f'  Our camera distortion coefficients:')
print(f'    k1 = {k1:+.6f}  (positive → barrel distortion)')
print(f'    k2 = {k2:+.6f}  (correction term)')
print(f'    fx = {fx_val:.2f} px  (focal length)')
print()

# Calculate distortion factor at different radii
print('  Distortion factor at different distances from centre:')
print(f'  {"Radius (px)":<15} {"r² term":<12} {"Scale Factor":<15} {"Error %"}')
print(f'  {"-"*55}')
for r_px in [0, 200, 500, 1000, 1500, 2000]:
    r_norm = r_px / fx_val
    r2 = r_norm ** 2
    scale = 1 + k1*r2 + k2*r2**2
    error_pct = abs(1 - scale) * 100
    print(f'  {r_px:<15} {r2:<12.4f} {scale:<15.6f} {error_pct:.2f}%')
print('=' * 60)
````
